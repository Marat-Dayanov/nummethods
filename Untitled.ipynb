```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 2 * x ** 4 - x ** 2 - 10

def f1(x):
    return 8 * x ** 3 - 2 * x
```

```python
x = np.linspace(-1.7, 1.7, 100)
y = f(x)
plt.figure(1)
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.grid()
plt.plot(x, y)
plt.show()
```

```python
di = 0.2
intervals = []
for i in np.arange(-2, 2, di):
    if f(i) * f(i + di) <= 0:
        intervals.append((i, i + di))
```

```python
def simple_iteration(a, b, fi, eps):
    x = (a + b) / 2
    count = 0
    while abs(f(x)) >= eps:
        x = fi(x)
        count += 1
    return x, count
```

```python
for start, end in intervals:
    c = -2/f1(start) / 2
    fi = lambda x: x + c * f(x)
    ans, count = simple_iteration(start, end, fi, 0.0001)
    print(ans)
    print(f(ans))
    print(count)
```
