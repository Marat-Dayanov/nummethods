```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x * x * np.exp(-x)
```

```python
a = 0.5
b = 1
eps = 1e-6
```

```python
n = 1
prev = 0
while True:
    h = (b - a) / n
    x = np.array([a + i * h for i in range(n + 1)])
    y = f(x)
    I = (np.sum(y) - y[0] / 2 - y[n] / 2) * h
    if abs(I - prev) < eps:
        break
    n += 1
    prev = I
```
