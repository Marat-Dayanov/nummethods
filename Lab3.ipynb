```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 1 / np.sqrt(x * (1 - x))
```

```python
def I(a, b):
    n = 1
    prev = 0
    eps = 1e-6
    while True:
        h = (b - a) / n
        x = np.array([a + i * h for i in range(n + 1)])
        y = f(x)
        I = (np.sum(y) - y[0] / 2 - y[n] / 2) * h
        if abs(I - prev) < eps:
            break
        n += 1
        prev = I
    return prev, n
```

```python
def calculate_f():
    a = 0.1
    b = 0.5
    eps = 1e-6
    da = 0.0001
    prev = 0
    while a > 0:
        i, n = I(a, b)
        if abs(i - prev) < eps:
            break
        a -= da
        prev = i
    return prev, n, a + da, b
res, n, a, b = calculate_f()
```

```python
def g(x):
    return x * np.exp(-x) / (x + 2)
```

```python
X = np.linspace(0, 10, 1000)
Y = g(X)
plt.figure(1)
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.fill_between(X, 0, Y, color="green")
plt.grid()
plt.plot(X, Y)
plt.show()
```
