```python
import matplotlib.pyplot as plt
import numpy as np
from math import log
```

```python
def f(x):
    return x + np.log(1 + x) / np.log(10) - 1.5

def f1(x):
    return 1 + 1 / (1 + x) / log(10)

def f2(x):
    return -1 / log(10) / (1 + x) / (1 + x)
```

```python
x = np.linspace(-3, 3, 100)
y = f(x)
plt.figure(1)
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.grid()
plt.plot(x, y)
plt.show()
```

```python
di = 0.2
for i in np.arange(-3, 3, di):
    if f(i) * f(i + di) <= 0:
        print(i, i + di)
```

```python
a = 1
b = 1.2
print(f(a) * f2(a) > 0)
print(f(b) * f2(b) > 0)
```

```python
def modified_newtons_metod(start, eps):
    k = f1(start)
    x = start
    count = 0
    while abs(f(x)) >= eps:
        x = x - f(x) / k
        count += 1
    return float(x), count
```

```python
ans, count = modified_newtons_metod(a, 0.00001)
print(f"{ans:.8f}")
print(count)
```
